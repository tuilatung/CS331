```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def compute_grad(theta, opt='theta1'):
    if opt == 'theta1':
        return 8*theta[1] - 8
    if opt == 'theta0':
        return 4*theta[0] - 8
```

```python
def compute_loss(theta):
    # theta is a two-dimensional vector 
    return 2 * theta[0]**2 + 4 * theta[1]**2 - 8 * theta[0] - 8 * theta[1] + 25
```

```python
alpha = 1e-3
epsilon = 1e-7
theta = [0, 0]
```

```python
history = []
while True:
    theta[1] = theta[1] - alpha * compute_grad(theta, opt='theta1')
    theta[0] = theta[0] - alpha * compute_grad(theta, opt='theta0')
    current_loss = compute_loss(theta)
    history.append(current_loss)
    print(f"Loss: {current_loss}")
    if abs(compute_grad(theta, opt='theta1')) < epsilon and abs(compute_grad(theta, opt='theta0')) < epsilon:
        break
```

```python
print('Min loss: ', current_loss)
print(f'Theta0 = {theta[0]}, theta1 = {theta[1]}')
```

```python
plt.plot(history, 'bo')
plt.title('History of Loss')
plt.xlabel('Iterations')
plt.ylabel('Loss value')
```
